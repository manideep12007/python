```python
'''
Strings is data type that stores a sequence of characters enclosed by quotes
'''
name = "Johncena"
print(name)
print(type(name))
```

```python
# len - number of characters in string
word = "Hello World"
print(len(word))
```

```python
# concatenation - + - to join strings
word1 = 'Hello'
word2 = 'Python'
print(word1 + ' ' + word2)
```

```python
# escape sequence chars used to format strings
sentence = 'This is my string1\nWritten'
print(sentence)
# \n - next line
# \' - add ' 
# \t - add tab space
```

```python
#f-string - format strings 
name = 'mandy'
city = 'hyderabad'
print(f"{name} lives in {city} city.")
```

```python
# Indexing - character at each position / index
word = 'Python'
# first character
print(word[0])
# negative indexing
# last character
print(word[-1])
```

```python
# slicing - subpart of string [start: end+1: step]
sentence = 'I code in python'
print(sentence[1:10])
print(sentence[::-1])
```

```python
# string methods - returns new string
# replace - to replace and return 
str1 = 'I love Hyderabad'
str2 = str1.replace("Hyderabad","New York")
print(str2)
print(str1)
```

```python
# strip - removes whitspaces 
str1 = '   python world '
print(str1.strip())
print(str1.lstrip())
print(str1.rstrip())
```

```python
# split - into list 
str1 = 'Python World'
lst = str1.split(' ')
print(lst)
```

```python
# join - list to string
lst = ['I', 'code', 'in', 'python']
str1 = ' '.join(lst)
print(str1)
```

```python
# index - returns index of char
# find - returns char or -1
str1 = 'python world'
print(str1.index('n'))
print(str1.find('z'))
print(str1.rindex('o'))
```

```python
# lower, upper, title, capitalize, swapcase
str1 = 'python WORLD'
print(str1.lower())
print(str1.upper())
print(str1.capitalize())
print(str1.swapcase())
print(str1.title())
```

```python
# startswith , endswith 
str1 = 'Python'
print(str1.startswith('P'))
print(str1.endswith('D'))
```

```python
# count - returns number of repetations
str1 = 'Hello world of python world'
print(str1.count('o'))
print(str1.count('world'))
```

```python
# conditionals 
# if-elif-else 
# if true do this else do this 
# executing set of code instructions based on conditions

age = 24
if age >= 18:
    print("You can cast vote")
else:
    print("You cannot vote")
```

```python
number = 56
if number % 2 == 1:
    print("Odd Number")
else:
    print('Even number')
```

```python
num = 7
if num % 2 == 0 and num % 3 == 0:
    print("multiple of 6")
elif num % 3 == 0:
    print("multiple of 3 not 2 and 6")
elif num % 2 == 0:
    print("multiple of 2 not 3 and 6")
else:
    print("not multiple of 2, 3 and 6")
```

```python
light = 'yellow'
if light == 'red':
    print('Stop')
elif light == 'yellow':
    print("Wait")
elif light == 'green':
    print('go')
else:
    print("light-broken")
```

```python
a, b = 12, 45
result = a if a > b else b 
print(result, 'was greatest number')
```

```python
a, b, c = 23,4,22
result = a if a > b and a > c else b if b > c else c
print(result, 'was greatest number')
```

```python
num = -45
result = 'postive' if num > 0 else 'negative' if num < 0 else 'zero'
print(result)
```
