```python
# keys are distint (no duplicates | unique)
# keys are immutable objects
# values are mutable objects 
d1 = {}
d1['firstname'] = 'john'
print(d1.get('firstname'))
# update - metod to overwrite or add new key-value 
d1.update({'lastname': 'cena'})
print(d1)
```

```python
# to remove values 
d1 = {
    'firsname': 'john', 'lastname': 'cena', 'city': 'new york'
}
print(d1.pop('city')) # removes and return value
print(d1.popitem()) # removes and return pair
print(d1)
d1.clear()
print(d1)
```

```python
# keys, values and items 
d1 = {
    'key1': 'value1',
    'key2': 'value2',
    'key3': 'value3'
}
print(d1.keys())
print(d1.values())
print(d1.items())
print(list(d1.values()))
print(len(d1.keys()))
```

```python
lst = [1,2,3]
d1 = {}
d1.update(enumerate(lst,start=100))
print(d1)
```

```python
# sets - collection of unique items
# sets are mutable 
# set items are immutable
# no duplicates
# True == 1 == 1.0

s1 = {1,2,3,5,True, 1.0,5.0}
print(s1)
```

```python
# methods
# add - add new element
# update - adds multiple elements
s1 = {1,2,3}
s1.add(4)
s1.add(4)
s1.add(5)
print(s1)
s1.update([4,5,6,7])
print(s1)
```

```python
# discard - removes element or None
# remove - removes element
# pop - removes and returns
s1 = {1,2,3,4,5,6,7,8}
print(s1.pop())
s1.remove(4)
s1.discard(100)
print(s1)
# clear - remove all
s1.clear()
print(s1)
```

```python
# set operations 
# union - | - all elements only once
# intersection - & - common elements
# difference - - - remaining elements in first set 
# symmetric difference - ^ - uncommon elements
s1 = {1,2,3,4,5}
s2 = {3,4,5,6,7}
print(s1 | s2)
print(s1 & s2)
print(s1 - s2)
print(s1 ^ s2)
```

```python
# empty set 
s1 = set()
print(s1)
```

```python
# frozen set is immutable object 
s1 = frozenset({1,2,3})
print(s1)
print(type(s1))
```
