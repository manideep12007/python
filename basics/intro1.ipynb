Types of operators

```python
# Arithematic operators: +, -, *, /, //, %, **
a = 12
b = 45
print(a + b)
print(b - a)
print(a * 5)
print(b / 6) # floating division
print(b // 6) # true or int division
print(round(a ** 0.5,3)) # a power b 
print(b % 6) # remainder / modulus
```

```python
# Comparsion or relational operators:
#  <, >, <=, >=, !=, ==
a = 34
b = 23
print(a > b)
print(a >= b)
print(a <= b)
print(a < b)
print(b == '23')
print(a != b)
```

```python
# logical operators: and, or, not
a = 12
b = 56
print(a % 5 == 0 and b % 8 == 0)
print(a % 5 == 0 or b % 8 == 0)
print(not b % 7 == 0)
```

```python
# Assignment operators: =, +=, /=, -=, *=,..
a = 5
print(a)
a += 50
print(a)
a //= 10
print(a)
```

```python
# Type conversion - one type to another type | implict conversion
# Type casting - explict conversion
print(12 / 7) # type conversion 
print(int(12 / 7)) # type casting
```

```python
# input - takes user input values
# default as string datatype
name = input("What is your name: ")
year = int(input("What is your birth year: "))
age = 2025 - year
print(f"{name} was {age} years old.")
```

```python
# string to another datatypes
a = '123'
print(int(a))
print(float(a))
print(bool(a)) # True 
print(bool('')) # False
```

```python
# number to another type
a = 5600
print(float(a))
print(str(a))
print(bool(a)) # True
print(bool(0)) # False
```

```python
# bool to another type
print(int(True)) # 1
print(float(True)) # 1.0
print(str(True))
```
