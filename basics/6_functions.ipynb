```python
# default parmeters - value assigned default 
# positional paramters - series of arguments passed 
# keyword arguments - keyword=value arguments to function
# *args - variable length of positional args
# **kwargs - variable length of keyword args

def total(*nums):
    return sum(nums)

print(total(1,2,3))

def d1(**kwargs):
    for key in kwargs:
        if key == 'city':
            print(f"{kwargs[key].capitalize()} is beautiful city")
d1(city='hyderbad',country='india',num=12)

def sayHello(fname='john',lname='cena'):
    return f'Hello, {fname} {lname}'

print(sayHello('kohli','virat'))
print(sayHello(fname='virat',lname='kohli'))
print(sayHello())
print(sayHello(lname='wick'))
```

```python
# Scope - region where variables and accessed
# python follows LEGB order
# local - defined inside function & accessed inside function itself
# enclosing - defined inside inner function 
# global - outside function and accessed everywhere
# built-in - variables as args to built-in functions

# global keyword used for modify global var inside local scope
# nonlocal keyword use of modify as local variable inside enclosing scope

x = 10 # global

def outer():
    y = 5 # local
    def inner():
        z = 3 # enclosing
        print(x,y,z)
    inner()
outer()

count = 0
def increment():
    global count
    count += 1
increment()
print(count)

def outer1():
    num = 10
    def inner():
        nonlocal num 
        num += 10
        print(num)
    inner()
    print(num)
outer1()
```

```python
# clousre - when inner function remembers variables from its enclosing scope, even after outer function has finished

def out(x):
    def inner(y):
        return x * y 
    return inner
f = out(5)
print(f(4)) # 20
```

```python
def counter():
    x = 0
    def inner():
        nonlocal x
        x += 1
        return x 
    return inner
count = counter()
print(count())
print(count())
```

```python
def prime(num):
    if num < 2:
        return False
    for i in range(2,int(num ** 0.5) + 1):
        if num % i == 0:
            return False
    return True

for el in range(1,25):
    if prime(el):
        print(el, end=' ')
```

```python
# recursive function - function calling itself again and again
def factorial(n):
    if n == 0:
        return 1
    for i in range(1,n+1):
        return n * factorial(n-1)
print(factorial(5))
print(factorial(6))
```

```python
# Count the number of vowels in a string
def count_vowels(string: str):
    count = 0
    for char in string:
        if char.isalpha() and char.lower() in 'aeiou':
            count += 1
    return count
print(count_vowels('manideep'))
```

```python
# Reverse a given string.
def reverse_string(strng: str) -> str :
    reverse = ''
    for char in strng:
        reverse = char + reverse
    return reverse
print(reverse_string('manideep'))
```

```python
# Print all even numbers between 1 and 50.
def even_nums(a,b):
    lst = []
    for i in range(a,b+1):
        lst.append(i) if i % 2 == 0 else None
    return lst 
print(even_nums(1,11))
```

```python
# Find the largest element in a list.
def largest_ele(lst):
    largest = 0
    for ele in lst:
        if ele > largest:
            largest = ele
            index = lst.index(ele)
    return largest,index
print(largest_ele([1,2,3,45,56,-3,56])) 
```

```python
# Remove duplicate elements from a list.
def remove_duplicates(lst):
    res = []
    for ele in lst:
        if ele not in res:
            res.append(ele)
    return res 
print(remove_duplicates([1,2,3,3,2,1]))
```

```python
# Count the frequency of each character in a string.
def frequency(word: str):
    vowels_freq = {}
    consoants_freq = {}
    specialchars_freq = {}
    numbers_freq = {}
    for char in word:
        if char.isalpha():
            if char.lower() in 'aeiou':
                if char in vowels_freq:
                    vowels_freq[char] += 1
                else:
                    vowels_freq[char] = 1
            else:
                if char in consoants_freq:
                    consoants_freq[char] += 1
                else:
                    consoants_freq[char] = 1
        elif char.isdigit():
            if char in numbers_freq:
                numbers_freq[char] += 1
            else:
                numbers_freq[char] = 1
        else:
            if char not in specialchars_freq:
                specialchars_freq[char] = 1
            else:
                specialchars_freq[char] += 1
    return vowels_freq,consoants_freq,specialchars_freq, numbers_freq

vowels,consonants,specialchars,numbers = frequency('hello python3')
print(vowels)
print(consonants)
print(specialchars)
print(numbers)
```

```python
# Write a function to check whether a number is prime.
def primelist(a,b):
    lowest = min(a,b)
    highest = max(a,b)
    for num in range(lowest,highest+1):
        isprime = True
        if num <= 1:
            isprime = False
        for i in range(2, int(num**0.5) + 1):
            if num % i == 0:
                isprime = False
                break
        if isprime:
            print(num,end=' ')
primelist(a=310,b=280)
```

```python
# Find the second largest number in a list
def second_largest(lst):
    n = len(lst)
    for i in range(n):
        for j in range(0,n-1):
            if lst[j] > lst[j+1]:
                lst[j],lst[j+1] = lst[j+1],lst[j]
    return lst[1]
print(second_largest([1,24,56,23,23]))
```

```python
# Write a function that returns another function to add a fixed number (closure)

def summation(x):
    def result(y):
        return x + y 
    return result
number1 = summation(10)
print(number1(10))
```

```python
def palindrome_string(string: str) -> bool:
    if len(string) == 1:
        return True
    reverse = ''
    for char in string:
        reverse = char + reverse
    return reverse == string 
print(palindrome_string('racecar'))
```

```python
def armstrong(num: float) -> bool :
    temp = num 
    count = 0
    while temp != 0:
        count += 1
        temp //= 10
    temp = num 
    summation = 0
    while temp != 0:
        digit = temp % 10
        summation += (digit ** count)
        temp //= 10
    return summation == num
print(armstrong(153))
print(armstrong(153))
for num in range(100,15000):
    if armstrong(num):
        print(num,end=' ')
```
