```python
# for loop
# used for finite number of iterations 

# while loop
# used for unknown number of iterations

# while condition:
#       code 

i = 1
while i <= 5:
    print(f'iteration{i}')
    i += 1
```

```python
i = 1
while i <= 3:
    print('Hello',end=' ')
    i += 1
```

```python
# loop statements
# break - terminates loop further iterations
# continue - terminates current iteration
# pass - do nothing 
i = 1
while i <= 10:
    if i % 2 == 0:
        i += 1
        continue
    print(i,end=' ')
    i += 1
```

```python
i = 1
while i <= 100:
    if i % 11 == 0:
        print(i,end=' ')
    i += 1
```

```python
i = 100
while i > 0:
    if i % 7 == 0:
        print(i,end=' ')
    i -= 1
```

```python
i = 5
while i > 0:
    j = 1
    while j <= 2:
        print(f'{i} X {j} = {i * j}')
        j += 1
    print()
    i -= 1
```

```python
# loop over lists or sets or tuples
lst = ['ironman','batman','hulk','superman']
i = 0
while i < len(lst):
    print(lst[i])
    i += 1
```

```python
# loop over dictonary
d1 = {
    'name': 'mandy',
    'age': 25,
    'city': 'hyderabad'
}
k = list(d1.keys())
i = 0
while i < len(k):
    print(k[i],d1[k[i]])
    i += 1
```

```python
lst = [1,2,34,68,5,34,45]
x = 3
i = 0
while i < len(lst):
    if lst[i] == x:
        print(i)
        break
    i += 1
else:
    print(-1)
```

```python
# for loops
for ele in 'python':
    print(ele.upper())
```

```python
lst = [1,2,3,4,5,6,8,8]
for ele in lst:
    print(ele ** 2)
```

```python
summation = 0
lst = [12,34,78,89,66,42]
for ele in lst:
    summation += ele
print(summation)
```

```python
num = 89
prime = True
if num > 2:
    for i in range(2,int(num ** 0.5) + 1):
        if num % i == 0:
            prime = False
            break
else:
    prime = False
if prime:
    print(num, 'is prime number')
else:
    print(num, 'is not prime number')
```

```python
for i in range(2,101):
    prime = True
    for j in range(2,int(i**0.5)+1):
        if i % j == 0:
            prime = False
            break
    if prime:
        print(i,end=' ')
else:
    print('\nthis are primes')
```

```python
freq = {}
word = 'Python Programming.'
for char in word:
    if char.isalnum() :
        if char not in freq:
            freq[char] = 1
        else:
            freq[char] += 1
print(freq)
```

```python
num = 6
fact = 1
if num == 0:
    print(fact)
else:
    for i in range(1,num+1):
        fact *= i
    print(fact)
```
