```python
# Lists and tuples are ordered collection 
# list is mutable object (changeable)
# tuple is immutable object (unchangeable)
```

```python
lst = [1,2,3,4,5]
heros = ['ironman','thor']
print(lst)
print(heros)
print(type(lst))
```

```python
# len - number of elements in list
lst = [1,2,3,5,34,24]
print(len(lst))
```

```python
# indexing and slicing
lst = ['ironman','thor','batman','supeman']
print(lst[0])
print(lst[-1])
print(lst[1:3])
print(lst[1::2])
```

```python
# list methods 
# to add element to list
lst = [1,2,3]
# append - adds at last
# insert - adds at index
lst.append(100)
lst.insert(0, 120)
lst[3] = 300
print(lst)
# extend - adds another list or multiple elements
lst.extend([7,8])
print(lst)
```

```python
# to remove elements
# pop - removes from last and returns
# remove - removes element
# clear - remove all elements
lst = [1,2,3,4,5,6,7]
print(lst)
print(lst.pop())
lst.remove(3)
print(lst)
lst.clear()
print(lst)
```

```python
# sort and reverse 
lst = [12,35,5,32,12]
lst.sort()
print(lst)
lst.sort(reverse=True)
print(lst)
lst = [12,34,56]
lst.reverse()
print(lst)
```

```python
# copy - copies elements
from copy import deepcopy
lst = [1,2,3]
lst1 = lst.copy() # shallow copy
print(lst is lst1) # False: returns new object 
lst2 = lst # references same object
print(lst is lst2) # True
lst1 = deepcopy(lst)
print(lst1)
```

```python
# reversed, sorted
# deep copy technique
# general methods on iterables
lst = [12,35,23,12]
lst1 = list(sorted(lst,reverse=True))
print(lst1)
print(lst)
print(lst is lst1) # False
```

```python
# list is hetrogenous object
# list support duplicates
# list is dynamic growable
# mutable object 
```

```python
# tuple is also ordered collection 
# tuple is immutable object
# tuple is hetrogenous object
# tuple allows duplicates
t = (1,2,4)
print(t)
print(type(t))
```

```python
t = 12,34,56,68
print(t[0])
print(t[1:])
print(t[::-1])
```

```python
# tuple methods 
t = 12,23,45,23
print(t.count(23))
print(t.index(23))
print(len(t))
```

```python
# sort and reverse the tuple 
t = 'abc', 'john','becky','brock','roman'
t1 = sorted(t,reverse=True)
print(t1)
t2 = sorted(t,reverse=False,key=len)
print(t2)
```
